```python
from gensim.models import Word2Vec
import pandas as pd
import nltk
print("All libraries imported successfully!")
```

```python
import os

# Read the text file

# Check if file exists and get the correct path
file_path = r"C:\Users\Suryaprakash\OneDrive\Documents\Python\word2Vec_prj\data\raw\sample_text.txt"

if os.path.exists(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        text = file.read()
else:
    print(f"File not found at: {file_path}")
    print("Available files in data directory:")
    data_dir = os.path.dirname(file_path)
    if os.path.exists(data_dir):
        print(os.listdir(data_dir))
    else:
        print(f"Data directory not found: {data_dir}")

print(text)
```

### tokenization - convertin sentences into double quotes represented words

```python
# Convert all text to lowercase
text = text.lower()
print(text)
```

### split the text into sentences and then into words

```python
# Split text into sentences
sentences = text.split("\n")
print(sentences)
```

### remove empty lines

```python
# sentences = [s.strip() for s in sentences if s.strip()]
# print(sentences)

sentences = [s.strip() for s in sentences if s.strip()]
print(sentences)
```

### remove Punctuation and tokenize

```python
import re 

tokenized_sentences = []
for sentence in sentences:
    tokens = re.sub(r'[^a-zA-Z\s]', '', sentence)
    words = tokens.split()

    tokenized_sentences.append(words)
print(tokenized_sentences)
```

### Train Your First Word2Vec Model


```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences = tokenized_sentences,
    vector_size = 100,
    window = 10,
    min_count = 1,
    workers = 4,
)
print("Word2Vec model trained successfully!")
```

### Check Vocabulary

```python
print(model.wv.index_to_key)
```

### Count Vocabulary Size

```python
print("Vocabulary Size:", len(model.wv.index_to_key))
```

### Inspect a Word Vector

```python
vector = model.wv['nvidia']

print(vector)
```

### Verify Vector Length

```python
print(len(model.wv['nvidia']))
```

### Find Similar Words

```python
print(
    model.wv.most_similar('nvidia')
)
```

### Compare Two Words

```python
print(
    model.wv.similarity(
        'nvidia',
        'developers'
    )
)
```

```python
words = model.wv.index_to_key

print("Total words:", len(words))
print(words[:20])
```

### Using PCA for 2d visualization of word embeddings

```python
from sklearn.decomposition import PCA

# Get vectors
vectors = model.wv[words]

# Reduce dimensions
pca = PCA(n_components=2)

result = pca.fit_transform(vectors)

print(result.shape)
```

### Plot the words

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

for i, word in enumerate(words):

    plt.scatter(
        result[i,0],
        result[i,1]
    )

    plt.annotate(
        word,
        xy=(result[i,0], result[i,1])
    )

plt.title("Word2Vec Embeddings - PCA")
plt.show()
```
