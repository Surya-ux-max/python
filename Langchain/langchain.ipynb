```python
text = '''
AI can personalize learning experiences for students.
AI can automate grading and feedback for teachers.'''
print(text)
```

```python
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader

splitter = CharacterTextSplitter(chunk_size=30, chunk_overlap=2)

chunks = splitter.split_text(text)
print(chunks)
```

```python
! pip install langchain-community langchain-huggingface
```

```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

db = FAISS.from_texts(
    texts=chunks,
    embedding=embedding_model
)
```

```python
results = db.similarity_search("What can AI do for students?", k=2)
print(results)
```
