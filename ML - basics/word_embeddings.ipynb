## Plotting Embeddings
lets now visulize some of our embeddings

### we are suing TSNE to reduce the 100 dimensional figure to 2 - Dimensions for the visualize because we cant plot the visualize for 100 words , we are finding the similar words for bert , it has 100 vectors though, TSNE will preserve the closed relationship of words.

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

target_word = "rnn"
selected_words = [w[0] for w in model.wv.most_similar(positive=[target_word], topn=200)] + [target_word]
embeddings = [model.wv[w] for w in selected_words] + model.wv[target_word]

mapped_embeddings = TSNE(n_components=2, metric='cosine',init='pca').fit_transform(embeddings)
```

```python
plt.figure(figsize=(20,20))
x = mapped_embeddings[:,0]
y = mapped_embeddings[:,1]
plt.scatter(x,y)

for i, txt in enumerate(selected_words):
    plt.annotate(txt, (x[i], y[i]))
```
