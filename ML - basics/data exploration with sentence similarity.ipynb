```python
import csv
import random
import os

# File path
f = r"C:\Users\Suryaprakash\OneDrive\Documents\Python\ML - basics\movie-dialog-corpus\movie_lines.tsv"

# Check if file exists
if not os.path.exists(f):
    print("File not found!")
    print("Current path:", f)
else:
    texts = []

    with open(f, "r", encoding="utf-8") as i:
        reader = csv.reader(i, delimiter="\t")

        for line in reader:
            if len(line) > 0:
                texts.append(line[-1])

    print("Total dialogues:", len(texts))

    # Shuffle data
    random.seed(1)
    random.shuffle(texts)

    # Take first 10000 dialogues
    texts = texts[:10000]

    print("\nFirst 5 dialogues:\n")
    for t in texts[:5]:
        print(t)
```

```python
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np

tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/2")


def embed_texts(texts):
    
    # Reduce logging output.
    tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)

    with tf.compat.v1.Session() as session:
        session.run([tf.compat.v1.global_variables_initializer(), tf.compat.v1.tables_initializer()])
        embeddings = session.run(embed(texts))

    return np.array(embeddings).tolist()

embeddings = embed_texts(texts)
```
